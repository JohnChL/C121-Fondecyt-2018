```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from dateutil.parser import parse
```

# CÁLCULO DE LA DEMANDA ENERGÉTICA REALIZADA PARA EL MÓDULO EXPERIMENTAL MEV1 – IMATA 

- La simulación para el cálculo de la demanda energética se realiza para los meses mayo, junio y julio del 2022.- El archivo de clima EPW se crea a partir de los datos de SENAMHI – Imata y Solcast.
- La temperatura neutra para cada mes se obtiene a partir de la ecuación establecida por la Dra. Jessica Molina. $$ Tn = 0.57 * Tm + 11.80 $$


# Resultados EnergyPlus

```python
archivo = '../IDF/idf_AFN/MEV1.csv'
```

```python
columnas = ['tiempo',"To","atico_cen","atico_E","modulo","atico_O"]

eplus = pd.read_csv(archivo,skiprows=1,names=columnas)
eplus.tiempo = eplus.tiempo.str.strip()
eplus.tiempo = "2022-" + eplus.tiempo
eplus.tiempo = eplus.tiempo.str.replace("24:00:00","23:59:00")
eplus.tiempo = pd.to_datetime(eplus.tiempo,format="%Y-%m/%d %H:%M:%S")
eplus.set_index("tiempo",inplace=True)
eplus
```

# Data experimental (HOBO)

```python
f = '../Data/002_Processed/MEV1_2022_2023.csv'
hobo = pd.read_csv(f, index_col=0, parse_dates=True)
hobo
```

# Gráficas

* To - temperatura ambiente
* Ti_sim - temperatura del interior simulada
* Ti_exp - temperatura del interior registrada

```python
fig, ax = plt.subplots(figsize = (12,6))
fecha1 = parse('2022-04-15')
fecha2 = fecha1 + pd.Timedelta('150D')

ax.plot(eplus.To, 'darkorange' , label = 'To')
ax.plot(eplus.modulo, 'steelblue', label = 'Ti_sim')
ax.plot(hobo.T_int1, 'k-', label = 'Ti_exp')


ax.set_xlim(fecha1, fecha2)
#ax.set_ylim(5, 12)

plt.xlabel('Fecha')
plt.ylabel('Temperatura °C')
ax.legend()
```

# Demanda energética

```python
archivo = '../IDF/idf_EnergReq/MEV1.csv'
```

```python
columnas = ['tiempo',"To","atico_cen","atico_E","modulo","atico_O",'Eatic_cen','Eatic_E','E_modulo','Eatic_O']

hvact = pd.read_csv(archivo,skiprows=1,names=columnas)
hvact.tiempo = hvact.tiempo.str.strip()
hvact.tiempo = "2022-" + hvact.tiempo
hvact.tiempo = hvact.tiempo.str.replace("24:00:00","23:59:00")
hvact.tiempo = pd.to_datetime(hvact.tiempo,format="%Y-%m/%d %H:%M:%S")
hvact.set_index("tiempo",inplace=True)
hvact
```

```python
total_energy= hvact['E_modulo'].groupby(by = [hvact['E_modulo'].index.month]).sum()*2.77778E-7
total_energy
## Considerar solo los meses 05, 06 y 07
```
