# An'alisis grafica

## Disponibilidad de datos sengun los datalogger HOBO

```python
mev1.plot(subplots = True, figsize = (12,25))
```

```python
mev1["T_int1"].loc["2021-12-20":"2022-08-20"].plot()
```

```python
fig, ax = plt.subplots(figsize = (12,6))
fecha1 = parse('2021-12-19')
fecha2 = fecha1 + pd.Timedelta('245D')

ax.plot(mev1["T_int1"], 'darkorange' , label = 'T_int')
ax.plot(senamhi.To, 'steelblue', label = 'T_amb')

ax.set_xlim(fecha1, fecha2)
# ax.set_ylim()

plt.xlabel('Fecha')
plt.ylabel('Temperatura °C')
plt. grid()
ax.legend()

plt.savefig("../imagenes/exported/MEV1_To_Tint_2022.png", dpi = 200)
```

```python
fig, ax = plt.subplots(figsize = (12,6))
fecha1 = parse('2022-02-01')
fecha2 = fecha1 + pd.Timedelta('2D')

ax.plot(mev1["T_int1"], 'darkorange' , label = 'T_int')
ax.plot(senamhi.To, 'steelblue', label = 'T_amb')

ax.set_xlim(fecha1, fecha2)
# ax.set_ylim()

plt.xlabel('Fecha')
plt.ylabel('Temperatura °C')
plt. grid()
ax.legend()

plt.savefig("../imagenes/exported/MEV1_To_Tint_02.2022.png", dpi = 200)
```
