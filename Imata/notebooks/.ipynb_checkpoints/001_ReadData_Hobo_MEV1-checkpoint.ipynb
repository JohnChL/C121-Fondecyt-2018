# Importamos la data [Raw] de las HOBO para el MEV1

```python
import pandas as pd
import numpy as np
```

## 1. Datalogger Hobo1

```python
f = '../Data/001_Raw/MEVs/MEV1/C_Paredes_exteriores.csv'
```

```python
df1 = pd.read_csv(f, index_col = 0, parse_dates=True)
df1.head()
```

```python
df = df1.rename_axis(index='Date')
```

```python
hobo1 = df.rename({'Temp, °C (SEN S/N: 20486630, LBL: Pared este)':'P_este',
                   'Temp, °C (SEN S/N: 20486630, LBL: Pared sur)':'P_sur',
                   'Temp, °C (SEN S/N: 20486630, LBL: Pared oeste)':'P_oeste',
                   'Temp, °C (SEN S/N: 20486630, LBL: Pared norte)':'P_norte'}, axis = 1)
hobo1
```

## Informacion sobre el Dataframe (DF)

```python
hobo1.info()
```

```python
hobo1.describe()
```

## Graficas

```python
hobo1.plot(figsize = (10,5))
```

## 2. Microestacion Hobo

```python
f2 = '../Data/001_Raw/MEVs/MEV1/Ti_HRi_Ilum_Ti.csv'
```

```python
df2 = pd.read_csv(f2, index_col = 0, parse_dates=True)
df2.head()
```

```python
df = df2.rename_axis(index='Date')
```

```python
hoboME = df.rename({'Temp, °C (SEN S/N: 2420039, LBL: Tint1)':'T_int1',
                   'HR, % (SEN S/N: 2420039, LBL: HR)':'RH',
                   'Intensidad, lux (SEN S/N: 2420039, LBL: I)':'lux',
                   'Temp, °C (SEN S/N: 2420039, LBL: S12_atico?)':'T_atico'}, axis = 1)

hoboME
```

## Informacion sobre el DF

```python
hoboME.info()
```

```python
hoboME.describe()
```

## Graficas

```python
hoboME.plot(figsize = (10,5))
```

```python
hoboME[["T_atico","T_int1"]].plot()
```

```python
hoboME.T_int1.plot(figsize = (10,5))
```

## Concat

```python
MEV1_2023 = pd.concat([hobo1,hoboME], axis=1)
MEV1_2023
```
