# Importando la data [raw] del SCS - Kallapuma

```python
import pandas as pd
```

## 1. Datalogger Hobo1

```python
f = '../data/001_Raw/SCS_kallapuma/Kallapuma_1_17.07.2023_10.51am.csv'
```

```python
df1 = pd.read_csv(f, index_col=0, parse_dates=True)
df1.head()
```

```python
df1 = df1.rename_axis(index='Date')
```

```python
hobo1 = df1.rename({'Temp, °C (SEN S/N: 20664779, LBL: Pared Sur)':'P_sur',
                   'Temp, °C (SEN S/N: 20664779, LBL: Pared Oeste)':'P_oeste',
                   'Temp, °C (SEN S/N: 20664779, LBL: Ventana)':'ventana',
                   'Temp, °C (SEN S/N: 20664779, LBL: Piso)':'piso'}, axis = 1)
hobo1
```

## Informacion sobre el DF

```python
hobo1.info()
```

```python
hobo1.describe()
```

## Graficas

```python
hobo1.plot()
```

```python
hobo1.plot(subplots = True)
```
