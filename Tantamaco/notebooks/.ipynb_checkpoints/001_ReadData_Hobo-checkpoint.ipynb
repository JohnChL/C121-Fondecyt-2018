# Importando la data [raw] del SCS - Tantamaco

```python
import pandas as pd
```

## 1. Datalogger Hobo1

```python
f = '../data/001_Raw/SCS_tantamaco/Tantamaco_Posta_1_13.07.23_9.27am.csv'
```

```python
df1 = pd.read_csv(f, index_col = 0, parse_dates=True)
df1.head()
```

```python
df = df1.rename_axis(index='Date')
```

```python
hobo1 = df.rename({'Temp, °C (SEN S/N: 20486631, LBL: Pared Sur)':'P_sur',
                   'Temp, °C (SEN S/N: 20486631, LBL: Psred Oeste)':'P_oeste',
                   'Temp, °C (SEN S/N: 20486631, LBL: Pared Oeste exterior)':'P_ext_oeste',
                   'Temp, °C (SEN S/N: 20486631, LBL: Ventana)':'ventana'}, axis = 1)
hobo1
```

## informacion sobre el DF

```python
hobo1.info()
```

```python
hobo1.describe()
```

## Graficas 

```python
hobo1.plot()
```

```python
hobo1.plot(subplots = True)
```
