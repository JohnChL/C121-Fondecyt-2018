# Importando la data [raw] del SCS - Tantamaco

```python
import pandas as pd
import glob
```

## 1. Datalogger Hobo1

```python
def import_hobo_1_tantamaco(f):
    df = pd.read_csv(f, index_col = 0, parse_dates=True)
    df = df.rename_axis(index='Date')
    
    df = df.rename({'Temp, °C (SEN S/N: 20486631, LBL: Pared Sur)':'P_sur',
                       'Temp, °C (SEN S/N: 20486631, LBL: Psred Oeste)':'P_oeste',
                       'Temp, °C (SEN S/N: 20486631, LBL: Pared Oeste exterior)':'P_ext_oeste',
                       'Temp, °C (SEN S/N: 20486631, LBL: Ventana)':'ventana'}, axis = 1)
    return df
```

```python
archivos = glob.glob("../data/001_Raw/Hobo/Tantamaco_Posta_1*.csv")
```

```python
lista = []
for archivo in archivos:
    lista.append(import_hobo_1_tantamaco(archivo))

    hobo1 = pd.concat(lista, axis=0)
```

```python
hobo1.sort_index(inplace = True)
```

```python
hobo1
```

```python
hobo1.info()
```

```python
hobo1.plot(subplots = True)
```

## 3. Microestacion

```python
def import_hobo_ME_tantamaco(f):
    df = pd.read_csv(f, index_col = 0, parse_dates=True)
    df = df.rename_axis(index='Date')
    
    df = df.rename({'Temp, °C (SEN S/N: 20658680, LBL: Tint)':'T_int',
                   'HR, % (SEN S/N: 20658680, LBL: HR)':'HR',
                   'Intensidad, lux (SEN S/N: 20658680, LBL: Luz)':'Luz'}, axis = 1)
    return df
```

```python
archivos = glob.glob("../data/001_Raw/Hobo/Tantamaco_ME*.csv")
```

```python
lista = []
for archivo in archivos:
    lista.append(import_hobo_ME_tantamaco(archivo))

    ME = pd.concat(lista, axis=0)
```

```python
ME.sort_index(inplace = True)
```

```python
ME
```

```python
ME.info()
```

```python
ME.plot(subplots = True)
```
